## Код

```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(2023)
```

```python
# Задаем сложный и зашумленный сигнал
time = np.linspace(0,60,1000)
signal = np.cos(2*time) + 2*np.cos(6*time) + 0.5*np.sin(12*time) + np.random.normal(0, 1,1000)
```

```python
fig, ax = plt.subplots(1,1,figsize=(8,6))

ax.plot(time, signal)

ax.grid()
ax.set_xlabel("t")
ax.set_ylabel("A")
ax.set_title("Original signal")
plt.plot()
```

```python
signal_fft = np.fft.fft(signal)
freq = np.fft.fftfreq(time.shape[-1])
```

```python
fig, ax = plt.subplots(1,1,figsize=(8,6))

ax.plot(freq, np.abs(signal_fft))
ax.set_xlim([0,0.2])

ax.grid()
ax.set_xlabel("frequency")
ax.set_ylabel("amplitude")
ax.set_title("Fourier transform")
plt.plot()
```

По поводу поиска частот. Не знаю, что конкретно нужно, но можно отсортировать массив по амплитудам и взять самые большие, сопоставив с графиком

```python
sorted([[amp, fr] for amp, fr in zip(np.abs(signal_fft), freq) if fr >= 0], reverse=True)[:10]
```
